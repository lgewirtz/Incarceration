#### Groupby year, avg the populations

```python
dfg = df.groupby(['year']).agg({'total_jail_pop_rate': ['mean'],
                                       'black_jail_pop_rate': ['mean'],
                                       'white_jail_pop_rate': ['mean'],
                                        'total_prison_pop_rate': ['mean'],
                                        'black_prison_pop_rate': ['mean'],
                                       'white_prison_pop_rate': ['mean']})
```

```python
dfg = dfg.reset_index()
```

```python
dfg.head(50)
```

```python
# delete 2017-2018 because too much missing information and looking for trends over time so 1990-2016 is fine

dfg2 = dfg.iloc[:27]
```

```python
dfg2.head(50)
```

```python
# Output data to file

dfg2.to_csv(os.path.join(path, 'Prepped Data', 'rates_group_year_avg.csv'))
```

```python
dfg2.columns
```

```python
dfg2.columns = ['year', 'tot_jail_pop_rate', 'black_jail_pop_rate', 'white_jail_pop_rate', 'total_prison_pop_rate', 'black_prison_pop_rate', 'white_prison_pop_rate']
```

```python
dfg2
```

#### Add prison and jail rates

```python
dfg2['tot_imprisoned'] = dfg2['tot_jail_pop_rate'] + dfg2['total_prison_pop_rate']
```

```python
dfg2.head()
```

```python
# Drop jail and pop columns after combined

dfg2.drop(['tot_jail_pop_rate', 'total_prison_pop_rate'], axis=1, inplace=True) 
```

```python
dfg2.head()
```

```python
dfg2['tot_black_imprisoned'] = dfg2['black_jail_pop_rate'] + dfg2['black_prison_pop_rate']
```

```python
# Drop jail and pop columns after combined

dfg2.drop(['black_jail_pop_rate', 'black_prison_pop_rate'], axis=1, inplace=True) 
```

```python
dfg2.head()
```

```python
dfg2['tot_white_imprisoned'] = dfg2['white_jail_pop_rate'] + dfg2['white_prison_pop_rate']
```

```python
# Drop jail and pop columns after combined

dfg2.drop(['white_jail_pop_rate', 'white_prison_pop_rate'], axis=1, inplace=True) 
```
