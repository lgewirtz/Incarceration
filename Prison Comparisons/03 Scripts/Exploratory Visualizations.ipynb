# 2. Explore variable relationships

### Create correlation matrix heatmap

```python
dfx = df[['tot_imprisoned', 'tot_black_imprisoned', 'tot_white_imprisoned']]
```

```python
dfx.head()
```

```python
# Create empty subplot

f,ax = plt.subplots(figsize=(10,10))
```

```python
# Create the correlation heatmap 

corr = sns.heatmap(dfx.corr(), annot = True, ax = ax) 
```

```python
corr.figure
```

#### There is moderate negative correlation btwn the rate of black imprisoned population and the rate of the white imprisoned population  


### Create Correlation Scatterplots

```python
dfx.head(1)
```

```python
# Create a scatterplot for the "tot_imprisoned" and "tot_black_imprisoned" columns in seaborn

sns.lmplot(x = 'tot_imprisoned', y = 'tot_black_imprisoned', data = dfx)
```

#### slight downward trend with a lot of variablity

```python
# Create a scatterplot for the "tot_imprisoned" and "tot_white_imprisoned" columns 

sns.lmplot(x = 'tot_imprisoned', y = 'tot_white_imprisoned', data = dfx)
```

#### Positive strong correlation with little variance between tot_white_imprisoned people and total_imprisoned people meaning as the rate of total imprisonment goes up, the rate of white people's imprisonment follows closely. 

```python
sns.lmplot(x = 'tot_white_imprisoned', y = 'tot_black_imprisoned', data = dfx)
```

#### There is a moderate negative correlation white and black imprisonment but with a lot of variablity
